```python
#Rolling Arima Method
#Inputs: NASDAQ 
#outputs: RMSE of the forecasted data
#split the data into 70% training and 30% testing data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
#TSA from statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt

#Display and Plotting
import seaborn as sns
```

```python
#Get the data
import yfinance as yf
stock = yf.Ticker('^IXIC').history(start='1985-01-01', end='2018-08-01')
#stock
X = stock["Close"]
train = X[int(len(X)*0.7):]
history = [x for x in train]
test = X[-int(len(X)*0.3):]
predictions = []
for t in range(len(test)):
    model = sm.tsa.ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(test, predictions))
error
```
