```python
#Rolling Arima Method
#Inputs: NASDAQ 
#outputs: RMSE of the forecasted data
#split the data into 70% training and 30% testing data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
#TSA from statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
#Display and Plotting
import seaborn as sns
```

```python
#Get the data
import yfinance as yf
stock1 = yf.Ticker('^IXIC').history(start='1985-01-01', end='2018-08-01')
stock2 = yf.Ticker('^N225').history(start='1985-01-01', end='2018-08-01')
stock3 = yf.Ticker('^DJI').history(start='1985-01-01', end='2018-08-01')
stock4 = yf.Ticker('^GSPC').history(start='1985-01-01', end='2018-08-01')
#stock

def arim(stock):
    X = stock["Close"]
    train = X[int(len(X)*0.7):]
    history = [x for x in train]
    test = X[-int(len(X)*0.3):]
    predictions = []
    for t in range(len(test)):
        model = sm.tsa.ARIMA(history, order=(5,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        history.append(obs)
        print('predicted=%f, expected=%f' % (yhat, obs))
    return test,predictions
```

```python
ixic_test,ixic_prediction = arim(stock1)
```

```python
n225_test,n225_prediction = arim(stock2)
```

```python
dji_test,dji_prediciton = arim(stock3)
```

```python
gspc_test,gspc_prediction = arim(stock4)
```

```python
from sklearn.metrics import mean_squared_error
ixic_error = np.sqrt(mean_squared_error(ixic_test, ixic_prediction))
n225_error = np.sqrt(mean_squared_error(n225_test, n225_prediction))
dji_error = np.sqrt(mean_squared_error(dji_test, dji_prediciton))
gspc_error = np.sqrt(mean_squared_error(gspc_test, gspc_prediction))
```

```python
print("IXIC RMSE: ",ixic_error)
print("N225 RMSE: ",n225_error)
print("DJI RMSE: ",dji_error)
print("GSPC RMSE: ",gspc_error)
```
